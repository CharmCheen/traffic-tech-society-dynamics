# 数据可视化

### 针对全球各国创新指数（GII）2024排名的可视化

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
file_path = 'D:\大学\observing\大计赛2025\数据集\平台数据转存\全球各国创新指数（GII）2024排名.csv'
df = pd.read_csv(file_path)

# 1. 各区域的国家数量分布
region_count = df['Region'].value_counts()

# 2. 不同区域的平均创新指数排名情况，保留两位小数
region_avg_rank = df.groupby('Region')['GII rank'].mean().round(2)

# 3. 查看创新指数排名前十的国家及其所属区域
top_10 = df.nsmallest(10, 'GII rank')[['Economy', 'Region']]

# 设置图片清晰度
plt.rcParams['figure.dpi'] = 300

# 1. 各区域的国家数量分布柱状图
plt.figure(figsize=(10, 6))
bars = plt.bar(region_count.index, region_count.values)
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.title('Number of Countries in Each Region')
plt.xticks(rotation=45)

# 添加数值标签
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')
plt.show()

# 2. 不同区域的平均创新指数排名情况柱状图
plt.figure(figsize=(10, 6))
bars = plt.bar(region_avg_rank.index, region_avg_rank.values)
plt.xlabel('Region')
plt.ylabel('Average GII Rank')
plt.title('Average GII Rank by Region')
plt.xticks(rotation=45)

# 添加数值标签
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')
plt.show()

# 3. 创新指数排名前十的国家及其所属区域柱状图
plt.figure(figsize=(10, 6))
bars = plt.bar(top_10['Economy'], range(1, 11))
plt.xlabel('Country')
plt.ylabel('GII Rank')
plt.title('Top 10 Countries by GII Rank and Their Regions')
plt.xticks(rotation=90)

# 添加数值标签
for bar, region in zip(bars, top_10['Region']):
    height = bar.get_height()
    plt.annotate(f'{region}', xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'D:\大学\observing\大计赛2025\数据集\平台数据转存\全球各国创新指数（GII）2024排名.csv'
df = pd.read_csv(file_path)

# 设置图片清晰度
plt.rcParams['figure.dpi'] = 300

# 绘制箱线图
bp = df.boxplot(column='GII rank', by='Region', patch_artist=True)

# 设置不同区域箱线图的颜色
colors = ['pink', 'lightblue', 'lightgreen', 'orange', 'yellow', 'purple', 'gray']
for patch, color in zip(bp.artists, colors):
    patch.set_facecolor(color)

# 创建图例
legend_handles = []
for i, region in enumerate(df['Region'].unique()):
    patch = plt.Rectangle((0, 0), 1, 1, fc=colors[i])
    legend_handles.append(patch)

plt.legend(legend_handles, df['Region'].unique(), loc='center left', bbox_to_anchor=(1, 0.5))

plt.xlabel('Region')
plt.ylabel('GII Rank')
plt.title('Box Plot of GII Rank by Region')
plt.suptitle('')  # 去除默认的副标题
plt.xticks(rotation=45)
plt.show()
```

通过对全球各国创新指数（GII）2024 排名数据的分析及可视化图表展示，可得出以下结论：  
区域创新能力差异显著：从平均创新指数排名来看，北美洲平均排名靠前，其中美国作为区域内的主要经济体，在科技创新领域投入巨大，在信息技术、生物科技等前沿领域成果丰硕，拉高了整个区域的排名。欧洲整体创新能力也较强，区域内多个国家如瑞士、瑞典、德国等长期在创新领域保持领先地位，完善的科研体系、充足的研发投入以及良好的创新文化氛围是其创新优势的重要支撑。而撒哈拉以南非洲平均排名靠后，该地区多数国家经济基础薄弱，在教育、科研投入等方面资源有限，导致创新能力发展受限。  
创新指数排名前十国家的区域分布集中：创新指数排名前十的国家中，欧洲占 7 个，这进一步凸显了欧洲在全球创新领域的核心地位。这些欧洲国家在高端制造业、精密仪器、航空航天等多个领域拥有深厚的技术积累和创新实力。东南亚、东亚和大洋洲有 2 个国家（新加坡和韩国）进入前十，这两个国家在电子信息、半导体、人工智能等新兴技术领域积极布局，凭借强大的产业集群和创新驱动的发展战略取得了突出的创新成果。北美洲的美国凭借其在科技研发、风险投资、高等教育等方面的绝对优势占据一席之地 。  
各区域国家数量分布不均：欧洲国家数量最多，这意味着在数据统计中其样本更为丰富，能更全面地反映该区域的创新情况。众多的国家和多样化的经济体系使得欧洲在创新发展上呈现出多元化的特点，不同国家在不同领域发挥着创新引领作用。北美洲国家数量最少，在分析该区域创新情况时，样本的局限性可能需要考虑，不过由于美国在创新领域的超强影响力，在一定程度上弥补了样本数量的不足。  
创新指数排名与区域发展水平相关：整体上，创新指数排名较高的区域，其经济发展水平、教育普及程度、基础设施建设等方面也相对更优。创新能力的提升有助于推动区域经济增长、产业升级，进而提高区域的整体发展水平；而良好的区域发展环境又为创新提供了更多的资源和支持，形成良性循环。例如，欧洲和北美洲的发达国家，凭借其发达的经济和优质的教育体系，能够吸引和培养大量的创新人才，投入充足的资金用于科研，从而保持较高的创新水平。  
区域内创新水平离散程度不同：通过箱线图可以看出，不同区域创新指数排名的离散程度存在差异。离散程度小的区域，内部国家的创新水平相对接近；离散程度大的区域，国家之间创新水平差距较大。这可能与区域内各国的经济结构、政策导向、资源禀赋等因素有关。例如，某些区域内既有创新能力极强的发达国家，也有创新基础薄弱的发展中国家，导致区域内创新水平离散程度较大。
