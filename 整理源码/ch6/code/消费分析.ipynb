```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

# 
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False

# Data
categories = [
    "", "/", "", 
    "", "", 
    "\n80", "\n50"
]

data = {
    'Feature': categories * 3,
    'Country': [''] * 7 + [''] * 7 + [''] * 7,
    '2023': [69, 63, 63, 53, 48, 49, 44, 77, 62, 53, 52, 41, 65, 60, 85, 87, 78, 90, 87, 95, 86],
    '2021': [65, 57, 52, 41, 36, 45, 40, 72, 55, 45, 37, 33, 57, 55, 83, 75, 66, 85, 85, 90, 83]
}

df = pd.DataFrame(data)

# Transform to long format
df_melted = df.melt(id_vars=['Feature', 'Country'], value_vars=['2021', '2023'], var_name='', value_name='')

# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(18, 10))
barplot = sns.barplot(
    data=df_melted,
    x='',
    y='Feature',
    hue='',
    ci=None,
    palette='Set2',
    edgecolor='black',
    dodge=True
)

plt.title('2021 vs 2023', fontsize=20)
plt.xlabel(' (%)', fontsize=14)
plt.ylabel('', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='', fontsize=12, title_fontsize=14, loc='lower right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm

# 
font_path = "/home/mw/input/ch491739173/Ubuntu_18.04_SimHei.ttf"
my_font = fm.FontProperties(fname=font_path, size=16)

# 
data = {
    '': [''] * 3 + [''] * 3 + [''] * 3,
    '': ['18-34', '35-54', '55+'] * 3,
    '': [80, 73, 69, 87, 89, 81, 85, 84, 77],
    '': [44, 44, 39, 57, 51, 41, 90, 93, 82],
    '': [35, 35, 35, 53, 52, 42, 93, 95, 81]
}

df = pd.DataFrame(data)

# 
car_types = ['', '', '']
colors = {'': '#FFCC00', '': '#3C3B6E', '': '#D62728'}

# 
fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

# 
for ax, car_type in zip(axs, car_types):
    age_groups = ['18-34', '35-54', '55+']
    x = np.arange(len(age_groups))
    total_width = 0.8
    single_width = total_width / 3
    offsets = [-single_width, 0, single_width]

    for i, country in enumerate(['', '', '']):
        values = df[df[''] == country][car_type].values
        bars = ax.bar(x + offsets[i], values, width=single_width, label=country, color=colors[country])
        for xi, yi in zip(x + offsets[i], values):
            ax.text(xi, yi + 1, f'{yi}%', ha='center', va='bottom', fontproperties=my_font, fontsize=12)

    ax.set_title(car_type, fontproperties=my_font, fontsize=16)
    ax.set_ylabel('%', fontproperties=my_font)
    ax.set_xticks(x)
    ax.set_xticklabels(age_groups, fontproperties=my_font)
    ax.set_ylim(0, 100)
    ax.grid(axis='y', linestyle='--', alpha=0.4)

# 
axs[0].legend(prop=my_font, loc='upper right')

# 
fig.suptitle('', fontproperties=my_font, fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
