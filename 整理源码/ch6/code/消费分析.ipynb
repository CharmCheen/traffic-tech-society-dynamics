```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

# 设置中文字体和允许负号显示
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False

# Data
categories = [
    "空调启动", "自适应头灯功能/性能", "座椅加热启动", 
    "交通拥堵辅助", "泊车辅助", 
    "更长的续航里程\n（如增加80公里）", "更大的发动机功率\n（如增加50马力）"
]

data = {
    'Feature': categories * 3,
    'Country': ['德国'] * 7 + ['美国'] * 7 + ['中国'] * 7,
    '2023': [69, 63, 63, 53, 48, 49, 44, 77, 62, 53, 52, 41, 65, 60, 85, 87, 78, 90, 87, 95, 86],
    '2021': [65, 57, 52, 41, 36, 45, 40, 72, 55, 45, 37, 33, 57, 55, 83, 75, 66, 85, 85, 90, 83]
}

df = pd.DataFrame(data)

# Transform to long format
df_melted = df.melt(id_vars=['Feature', 'Country'], value_vars=['2021', '2023'], var_name='年份', value_name='百分比')

# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(18, 10))
barplot = sns.barplot(
    data=df_melted,
    x='百分比',
    y='Feature',
    hue='年份',
    ci=None,
    palette='Set2',
    edgecolor='black',
    dodge=True
)

plt.title('不同国家用户功能偏好变化（2021 vs 2023）', fontsize=20)
plt.xlabel('百分比 (%)', fontsize=14)
plt.ylabel('功能项', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='年份', fontsize=12, title_fontsize=14, loc='lower right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm

# 指定中文字体
font_path = "/home/mw/input/ch491739173/Ubuntu_18.04_SimHei.ttf"
my_font = fm.FontProperties(fname=font_path, size=16)

# 数据
data = {
    '国家': ['德国'] * 3 + ['美国'] * 3 + ['中国'] * 3,
    '年龄段': ['18-34岁', '35-54岁', '55+岁'] * 3,
    '油车': [80, 73, 69, 87, 89, 81, 85, 84, 77],
    '插电式混合动力车': [44, 44, 39, 57, 51, 41, 90, 93, 82],
    '纯电动车': [35, 35, 35, 53, 52, 42, 93, 95, 81]
}

df = pd.DataFrame(data)

# 绘图参数
car_types = ['油车', '插电式混合动力车', '纯电动车']
colors = {'德国': '#FFCC00', '美国': '#3C3B6E', '中国': '#D62728'}

# 设置子图
fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

# 绘制每种动力类型的柱状图
for ax, car_type in zip(axs, car_types):
    age_groups = ['18-34岁', '35-54岁', '55+岁']
    x = np.arange(len(age_groups))
    total_width = 0.8
    single_width = total_width / 3
    offsets = [-single_width, 0, single_width]

    for i, country in enumerate(['德国', '美国', '中国']):
        values = df[df['国家'] == country][car_type].values
        bars = ax.bar(x + offsets[i], values, width=single_width, label=country, color=colors[country])
        for xi, yi in zip(x + offsets[i], values):
            ax.text(xi, yi + 1, f'{yi}%', ha='center', va='bottom', fontproperties=my_font, fontsize=12)

    ax.set_title(car_type, fontproperties=my_font, fontsize=16)
    ax.set_ylabel('受访者占比（%）', fontproperties=my_font)
    ax.set_xticks(x)
    ax.set_xticklabels(age_groups, fontproperties=my_font)
    ax.set_ylim(0, 100)
    ax.grid(axis='y', linestyle='--', alpha=0.4)

# 添加图例
axs[0].legend(prop=my_font, loc='upper right')

# 总标题
fig.suptitle('不同国家不同年龄段消费者对汽车动力系统类型的偏好', fontproperties=my_font, fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
