```
# 导入库
import pandas as pd
# 导入库
import matplotlib
matplotlib.use("TkAgg")
# 导入库
import matplotlib.pyplot as plt
# 导入库
import matplotlib.colors as mcolors

# 读取 Excel 文件
# 变量赋值
excel_file = pd.ExcelFile(r"C:\Users\x8615\Desktop\乱七八糟的文件\比赛\大计赛\第三章数据集\图2-13出行方式变革数据来源交通运输行业发展统计公报.xlsx")

# 获取指定工作表中的数据
# 变量赋值
df = excel_file.parse('Sheet1')

# 设置图片清晰度
# 变量赋值
plt.rcParams['figure.dpi'] = 300

# 设置中文字体
# 变量赋值
plt.rcParams['font.sans-serif'] = ['SimHei']
# 变量赋值
plt.rcParams['axes.unicode_minus'] = False

# 统一减小字体大小
plt.rcParams.update({'font.size': 6})

# 定义一组颜色循环，用于饼图
# 变量赋值
colors = list(mcolors.TABLEAU_COLORS.values())

# 创建一个 2x2 的子图布局
# 变量赋值
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# 遍历每一年的数据
for i, (index, row) in enumerate(df.iterrows()):
# 变量赋值
    year = int(row['年份'])  # 将年份转换为整数，去除小数点
# 变量赋值
    modes = df.columns[1:]
# 变量赋值
    values = row[modes]

    # 计算子图的位置
# 变量赋值
    row_idx = i // 2
# 变量赋值
    col_idx = i % 2
# 变量赋值
    wedges = axes[row_idx, col_idx].pie(values, autopct='%1.1f%%', textprops={'fontsize': 6}, colors=colors, radius=1)[0]
    # 缩小标题字体大小，这里设为 10，你可以根据需要调整
# 变量赋值
    axes[row_idx, col_idx].set_title(f'{year} 年各出行方式占比', fontweight='bold', fontsize=10)
    # 设置子图背景颜色为淡灰色
    axes[row_idx, col_idx].set_facecolor('#f9f9f9')

# 获取所有类别名称
# 变量赋值
modes = df.columns[1:]

# 设置全局图例
fig.legend(
    wedges, modes,
# 变量赋值
    loc='lower center',
# 变量赋值
    fontsize=6,
# 变量赋值
    ncol=len(modes),
# 变量赋值
    frameon=True,  # 显示图例边框
# 变量赋值
    facecolor='white',  # 图例背景颜色为白色
# 变量赋值
    edgecolor='gray',  # 图例边框颜色为灰色
# 变量赋值
    labelspacing=1.0  # 图例项之间的垂直间距
)

# 调整子图之间的间距
# 变量赋值
plt.subplots_adjust(wspace=0.3, hspace=0.3)

# 显示图表
plt.show()
```

```
# 导入库
import matplotlib

matplotlib.use("TkAgg")
# 导入库
import matplotlib.pyplot as plt
# 导入库
import numpy as np

# 设置图片清晰度
# 变量赋值
plt.rcParams['figure.dpi'] = 300

# 解决中文显示问题
# 变量赋值
plt.rcParams['font.sans-serif'] = ['SimHei']
# 变量赋值
plt.rcParams['axes.unicode_minus'] = False

# 数据来源：知网论文《交通强国背景下交通运输行业碳达峰碳中和路径研究》
# 变量赋值
years = ['2023年', '2025年', '2030年', '2035年', '2040年', '2045年', '2050年', '2055年', '2060年']
# 变量赋值
road_emission = [10, 11, 12, 13, 9, 7, 4, 2, 1]  # 公路碳排放
# 变量赋值
civil_aviation_emission = [1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]  # 民航碳排放
# 变量赋值
waterway_emission = [1.5, 1.6, 1.7, 1.8, 1.9, 2, 2.1, 2.2, 2.3]  # 水路碳排放
# 变量赋值
railway_emission = [0.8, 0.9, 1, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6]  # 铁路碳排放
# 变量赋值
total_emission = [12.99, 13.46, 14.59, 15.46, 12.79, 9.98, 6.48, 3.30, 1.27]  # 碳排放总量

# 变量赋值
bar_width = 0.2
# 变量赋值
index = np.arange(len(years))

# 定义不同的颜色
# 变量赋值
bar_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# 绘制柱状图，取消形状填充
# 变量赋值
plt.bar(index - 1.5 * bar_width, road_emission, width=bar_width, label='公路碳排放', color=bar_colors[0])
# 变量赋值
plt.bar(index - 0.5 * bar_width, civil_aviation_emission, width=bar_width, label='民航碳排放', color=bar_colors[1])
# 变量赋值
plt.bar(index + 0.5 * bar_width, waterway_emission, width=bar_width, label='水路碳排放', color=bar_colors[2])
# 变量赋值
plt.bar(index + 1.5 * bar_width, railway_emission, width=bar_width, label='铁路碳排放', color=bar_colors[3])

# 绘制折线图
# 变量赋值
plt.plot(index, total_emission, marker='o', label='碳排放总量', color='orange', linestyle='--')

# 定义垂直和水平偏移量
# 变量赋值
vertical_offset = 0.5
# 变量赋值
horizontal_offset = 0.1

# 在折线图上显示数据，调整位置
for i, val in enumerate(total_emission):
# 变量赋值
    plt.text(index[i] + horizontal_offset, val + vertical_offset, f'{val:.2f}', ha='center', va='bottom', fontsize=8,
# 变量赋值
             color='orange')

# 设置坐标轴标签和标题
# 变量赋值
plt.xticks(index, years, rotation=45)
plt.ylabel('交通碳排放总量/亿t')
plt.title('低碳情景下不同交通运输方式碳排放占比图')

# 显示图例
plt.legend()

# 添加数据来源说明
plt.text(0.5, -0.3, "数据来源：知网论文《交通强国背景下交通运输行业碳达峰碳中和路径研究》",
# 变量赋值
         transform=plt.gca().transAxes, ha='center', fontsize=8, color="grey")

# 调整布局
plt.tight_layout()

# 显示图表
plt.show()
```

```
# 导入库
import csv
# 导入库
import json
# 导入库
from DrissionPage import ChromiumPage

# 定义每页获取数据的函数
# 函数定义
def get_jobs_on_page(dp, page_num):
# 变量赋值
    url = f"https://www.zhipin.com/web/geek/job?query=交通&city=100010000&page={page_num}"
    dp.get(url)
# 变量赋值
    resp = dp.listen.wait()
# 变量赋值
    json_data = resp.response.body
    return json_data['zpData']['jobList']
# 定义总页数
# 变量赋值
total_pages = 10  # 你可以根据实际情况修改这个值

# 变量赋值
with open('data.csv', mode='w', encoding='utf-8', newline='') as f:
# 变量赋值
    csv_writer = csv.DictWriter(f, fieldnames=['职位'])
    csv_writer.writeheader()

# 变量赋值
    dp = ChromiumPage()
    dp.listen.start("zpgeek/search/joblist.json")

    for page_num in range(1, total_pages + 1):
# 变量赋值
        jobList = get_jobs_on_page(dp, page_num)
        for index in jobList:
# 变量赋值
            dit = {
                '职位': index['jobName']
            }
            csv_writer.writerow(dit)
            print(dit)
```

```
# 导入库
from wordcloud import WordCloud
# 导入库
import matplotlib

matplotlib.use("TkAgg")
# 导入库
import matplotlib.pyplot as plt
# 导入库
import pandas as pd

# 读取Excel文件
# 变量赋值
file_path = r"C:\Users\x8615\Desktop\乱七八糟的文件\比赛\大计赛\第三章数据集\数据来源职友集.xlsx"  # 替换为你的Excel文件路径
# 变量赋值
df = pd.read_excel(file_path)

# 提取传统岗位和新兴岗位数据
# 变量赋值
traditional_jobs = ' '.join(df['传统交通行业岗位'].dropna().tolist())
# 变量赋值
emerging_jobs = ' '.join(df['新兴岗位'].dropna().tolist())

# 设置Matplotlib字体配置
# 变量赋值
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
# 变量赋值
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号


# 生成词云图的函数
# 函数定义
def generate_wordcloud(text, title, font_path):
# 变量赋值
    wordcloud = WordCloud(
# 变量赋值
        width=400,  # 设置宽度和高度为相同值
# 变量赋值
        height=400,
# 变量赋值
        background_color='white',
# 变量赋值
        font_path=font_path,  # 指定中文字体路径
# 变量赋值
        relative_scaling=0.2  # 调整字体大小的均衡程度，值越接近0越均衡
    ).generate(text)

# 变量赋值
    plt.figure(figsize=(5, 5))  # 设置图形的宽高比为1:1
    plt.imshow(wordcloud)
    plt.axis('off')
    # 取消标题
    # plt.title(title)
    plt.show()


# 定义字体路径
# 变量赋值
font_path = r"C:/Windows/Fonts/simhei.ttf"

# 生成传统岗位词云图
generate_wordcloud(traditional_jobs, "传统交通行业岗位词云图", font_path)

# 生成新兴岗位词云图
generate_wordcloud(emerging_jobs, "新兴岗位词云图", font_path)
```

### 词云2.py

```
# 导入库
import pandas as pd
# 导入库
from wordcloud import WordCloud, STOPWORDS
# 导入库
import matplotlib
matplotlib.use("TkAgg")
# 导入库
import matplotlib.pyplot as plt

# 读取 CSV 文件
# 变量赋值
df = pd.read_csv('data.csv')  # 替换为你的文件路径

# 数据预处理（根据实际情况调整）
# 变量赋值
text = '\n'.join(df['职位'].dropna().astype(str))  # 替换为你的文本列名

# 生成词云
# 变量赋值
wordcloud = WordCloud(
# 变量赋值
    width=800,
# 变量赋值
    height=600,
# 变量赋值
    background_color='white',
# 变量赋值
    stopwords=STOPWORDS,
# 变量赋值
    max_words=200,
# 变量赋值
    font_path='simhei.ttf'  # 中文需要指定字体文件
).generate(text)

# 显示词云
# 变量赋值
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()

# 保存词云图片
wordcloud.to_file('wordcloud.png')
```
